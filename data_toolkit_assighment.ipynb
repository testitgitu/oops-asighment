1.  What is NumPy, and why is it widely used in Python?

    NumPy (Numerical Python) is a fundamental library for numerical computing in Python. It provides support for powerful N-dimensional arrays and offers a wide range of mathematical functions to operate on these arrays efficiently. NumPy is highly optimized and performs operations faster than native Python lists. It enables features like broadcasting and vectorization, which allow for concise and efficient code. Widely used in data science, machine learning, and scientific computing, NumPy serves as the foundation for many libraries such as pandas, scikit-learn, and TensorFlow. Its speed, ease of use, and versatility make it essential for numerical and analytical tasks in Python.

2.  How does broadcasting work in NumPy?

     Broadcasting in NumPy allows arrays of different shapes to be used together in arithmetic operations. When performing operations, NumPy automatically expands the smaller array to match the shape of the larger one without making actual copies of data. This works when the dimensions are either equal or one of them is 1. Broadcasting simplifies code and enhances performance by avoiding explicit loops or reshaping. For example, adding a one-dimensional array to a two-dimensional array applies the operation across rows or columns. It is widely used in data manipulation, mathematical computations, and machine learning tasks for its efficiency and flexibility.

3.  What is a Pandas DataFrame?

    A Pandas DataFrame is a two-dimensional, labeled data structure in Python, similar to a table or spreadsheet. It is part of the Pandas library and is widely used for data manipulation and analysis. It provides powerful tools for filtering, sorting, grouping, merging, and reshaping data. DataFrames are especially useful for working with structured data, such as CSV files or SQL tables. Their flexibility, ease of use, and integration with other libraries make them essential in data science.

6.    What are the differences between NumPy arrays and Python lists?

      NumPy arrays and Python lists differ in functionality, performance, and memory usage. NumPy arrays are **homogeneous, meaning all elements must be of the same data type, while Python lists can hold heterogeneous data types. NumPy arrays are faster and more memory-efficient because they use fixed-type data and optimized C code under the hood. They support vectorized operations, allowing element-wise computations without explicit loops, unlike lists which require iteration. NumPy also offers numerous built-in mathematical functions for operations on arrays. In contrast, Python lists are more flexible for general-purpose use but less efficient for numerical computing and large data sets.

8.   What does the term “vectorized operation” mean in NumPy?

     A vectorized operation in NumPy refers to performing element-wise computations on entire arrays without using explicit Python loops. Instead of processing each element one by one, NumPy applies operations directly to arrays using optimized, low-level code. This makes calculations faster and more efficient. For example, adding two arrays or multiplying all elements by a number can be done in a single step. Vectorization improves performance by leveraging hardware acceleration and reducing Python overhead. It also leads to cleaner, more readable code by replacing loops with simple expressions, making NumPy ideal for numerical and scientific computing tasks.

17.    Explain the difference between apply() and map() in Pandas?

        In Pandas, `apply()` and `map()` are used to manipulate data, but they differ in scope and functionality. `map()` is primarily used with Series and applies a function element-wise. It is ideal for simple operations like value transformation or replacement in a single column. In contrast, `apply()` works with both Series and DataFrames. It can apply a function along either axis (rows or columns) and is more flexible, supporting complex operations. While `map()` is limited to 1D structures, `apply()` is better suited for handling multi-dimensional data and more advanced computations across rows or columns in a DataFrame.

21.    Why is NumPy’s array slicing faster than Python’s list slicing?

       NumPy’s array slicing is faster than Python’s list slicing due to its **efficient memory management and optimized C-based implementation**. NumPy arrays are stored in **contiguous memory blocks**, allowing quick access to elements without the overhead of Python’s dynamic list structures. When slicing, NumPy creates a **view** of the original array rather than copying data, which saves time and memory. In contrast, Python lists are arrays of pointers to objects, and slicing involves creating a new list with copied references, making it slower. NumPy also benefits from **low-level optimizations and vectorized operations**, significantly boosting performance in numerical computations.

```python
#  How do you create a 2D NumPy array and calculate the sum of each row
import numpy as np

array_2d = np.array([[1, 2, 3],
                     [4, 6, 6],
                     [9, 10, 9]])

row_sums = np.sum(array_2d, axis=1)

print("2D Array:")
print(array_2d)

print("Sum of each row:")
print(row_sums)

```

```python
#  Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [20, 30, 30, 40],
    'Score': [85, 90, 88, 92]
}

df = pd.DataFrame(data)

mean_score = df['Score'].mean()
mean_Age = df['Age'].mean()

print("Mean of the 'Score' column:", mean_score)
print("Mean of the 'Age' column:", mean_Age)
```

```python
#  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create or load a DataFrame
data = {
    'Math': [90, 80, 70, 60, 50],
    'Science': [85, 75, 65, 55, 45],
    'English': [88, 78, 68, 58, 48]
}
df = pd.DataFrame(data)

# Step 2: Calculate the correlation matrix
correlation_matrix = df.corr()

# Step 3: Visualize with a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add a title
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()


```

```python
#  Perform matrix multiplication using NumPy
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Method 1: Using np.dot()
result_dot = np.dot(A, B)

# Method 2: Using @ operator
result_at = A @ B

# Display the results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result (using np.dot):\n", result_dot)
print("Result (using @ operator):\n", result_at)
```
